### Numerical Variables
Since the numerical variables are skewed we will perform log normal distribution

```python
data[numerical_features].columns
```

```python
numerical_features = ['Longitude', 'Latitude','Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']
```

```python
# Define the small constant
small_constant = 1e-9  # You can adjust this value based on your data characteristics

# Loop through each continuous feature
for feature in numerical_features:
    data = dataset.copy()
    if (data[feature] == 0).any():
        print(f"Applying logarithmic transformation with a small constant for {feature}")
        data[feature] = np.log(data[feature] + small_constant)
    else:
        data[feature] = np.log(data[feature])
    data['Aggregate rating'] = np.log(data['Aggregate rating'])
```

### Handling Rare Categorical Feature
Removing categorical variables which are present with less than 1% of the observations

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
categorical_features
```

```python
for feature in categorical_features:
    temp=dataset.groupby(feature)['Aggregate rating'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(100)
```

```python
for feature in categorical_features:
    labels_ordered=dataset.groupby([feature])['Aggregate rating'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
## for feature slection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) 
feature_sel_model.fit(X_train, y_train)
```
